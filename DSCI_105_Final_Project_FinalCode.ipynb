# Scrape information from a list of colleges, temporarily saving each college's information as an .html file

```python
list_test = ['https://www.usnews.com/best-colleges/princeton-university-2627',
             'https://www.usnews.com/best-colleges/university-of-california-berkeley-1312']
```

```python
from selenium import webdriver
import requests
from bs4 import BeautifulSoup

for college in list_test:
      
    print(college)
    
    # Using Selenium we grab the html code from each university and with BeautifulSoup we use the html.parser
    driver = webdriver.Chrome()
    driver.get(college)
    with open('temporary_file.html', "w", encoding='utf-8') as f:
        f.write(driver.page_source)
    
    # We then process each class that we are looking for and set it to a variable
    college_ranking_file = open("temporary_file.html", "r", encoding="utf8")  
    contents = college_ranking_file.read() 
    soup = BeautifulSoup(contents, 'html.parser')
    
    name_full = soup.find('h1', class_='Heading-sc-1w5xk2o-0 Wakanda__Title-rzha8s-10 gTHiRT gtkLHO').text
    print(name_full)

    # We use if statements to help search for incomplete data. Many schools were missing data points
    # After a certain amount of schools, USnews simply puts a group ranking, we take the lower number and assign it to each school
    ranking_text = soup.find('span', class_='Villain__RankingSpan-sc-8s66oj-4 fDSmVR').text
    ranking = ranking_text.split(" ")[0].replace('#', "")
    if "-" in ranking:
        ranking = ranking.split("-")[0]
    print("Ranked: " + ranking)
    
    text = soup.find_all('p', class_='Paragraph-sc-1iyax29-0 kqzCwD')
    
    setting = text[0].text
    undergraduate_enrollment = text[2].text.replace(",", "")
    acceptance_rate = text[3].text.replace("%", "")
    
    if "/" in text[4].text:
            sat_lower = text[4].text.split('-')[0]
            sat_upper = sat_lower
    else:
        sat_lower = text[4].text.split('-')[0]
        sat_upper = text[4].text.split('-')[1]

    if "/" in text[5].text:
        act_lower = text[5].text.split('-')[0]
        act_upper = act_lower
    else:
        act_lower = text[5].text.split('-')[0]
        act_upper = text[5].text.split('-')[1]
    
    hs_GPA = text[6].text
    
    # Public and Private schools are not listed the same for the below variables
    if (text[1].text == text[8].text):
        school_type = 'Public'
        in_state_cost = text[1].text.replace("$", "").replace(",", "")
        out_state_cost = text[7].text.replace("$", "").replace(",", "")
        private_cost = ''
        room_and_board = text[9].text.split("(")[0].replace("$", "").replace(",", "")
        avg_need_based_aid_package = text[10].text.replace("$", "").replace(",", "")
    else:
        school_type = 'Private'
        in_state_cost = ''
        out_state_cost = ''
        private_cost = text[1].text.replace("$", "").replace(",", "")
        room_and_board = text[8].text.split("(")[0].replace("$", "").replace(",", "")
        avg_need_based_aid_package = text[9].text.replace("$", "").replace(",", "")
        
    bibliometric_rank = text[-14].text
    graduate_enrollment = text[-9].text.replace(",", "")
    total_enrollment = text[-8].text.replace(",", "")
    full_time_students = text[-7].text.replace(",", "")
    graduation_rate = text[-6].text.replace("%", "")
    rate_graduating_students_who_have_borrowed = text[-5].text.replace("%", "")
    avg_debt_at_graduation = text[-4].text.replace("$", "").replace(",", "")
    
    print("Setting: " + setting)
    print("Type: " + school_type)
    print("Private Cost: " + private_cost)
    print("In-State Cost : " + in_state_cost)
    print("Out-State Cost: " + out_state_cost)
    print("Undergraduate Enrollment: " + undergraduate_enrollment)
    print("Acceptance Rate: " + acceptance_rate)
    print("SAT Range: " + sat_range)
    print("ACT Range: " + act_range)
    print("HS GPA: " + hs_GPA)
    print("Room & Board: " + room_and_board)
    print("Average Need Based Aid Package: " + avg_need_based_aid_package)
    print("Bibliometric Rank: " + bibliometric_rank)
    print("Graduate Enrollment: " + graduate_enrollment)
    print("Total Enrollment: " + total_enrollment)
    print("Full-Time Students: " + full_time_students)
    print("Graduation Rate: " + graduation_rate)
    print("Percentage of Graduating Students Who've Borrowed: " + rate_graduating_students_who_have_borrowed)
    print("Average Debt at Graduation: " + avg_debt_at_graduation)

    print()
    print("------------------------------------------------------------------------------------")
    print()
```

# Project Report

### Methods

   Using the selenium package, we went through each university link from the usnews top national universities (US). Selenium would gather the entire pages html, which we then used to scrape data from. Our biggest issue was the styling of the pages. USnews requires users to scroll towards the bottom of the page for more of it to load, this is why Selenium was vital to the project as it was able to bypass this feature and get the whole script. 
   
  Web scraping was broken into two sections. A heavy portion was in one class, and we used an indexing approach to gather the data. However, the remaining were in their own classes. Both situations took advantage of varying string manipulations, such as .text, .replace, and .strip. We did come across an issue with public schools. Since Selenium was slow, we had been using a sample test list. This test list was only private schools. Once we realized that public schools also had in-state and out-of-state tuition, we had to adjust our code to include this.
  
  Some data manipulation was the adding of longitude and latitude coordinates via the geocode package. We separated each school address and using only the zipcode, geocode was able to return the longitude and latitude of the location. This allowed for mapping opportunities in Tableau. We utilized if-else statements to leave objects blank when there was no data for a specific school. This allowed us to use Tableaus remove NA feature during analysis. This was crucial for the ethnicity rankings per school. Many schools near the bottom of the list only had a few ethnicity groups, rather than seven (the average). There were also a lot of testing code along the way to ensure the code is running properly, such as checking if the university was already completed or not.
